## nhập 1 tháng trong năm nhuận rồi in ra số ngày

```python
thang = int(input())
nam = int(input())
if thang in [1, 3, 5, 7, 8, 10, 12]:
    print(f"tháng {thang} có 31 ngày")
elif thang in [4, 6, 9, 11]:
    print(f"tháng {thang} có 30 ngày")
elif (nam % 4 == 0 and nam % 100 != 0) or (nam % 400 == 0):
    print(f"tháng {thang} có 29 ngày")
else:
    print(f"tháng {thang} có 28 ngày")
```

```python

nam = int(input())
thang = int(input())

match thang:
    case 1|3|5|7|8|10|12:
        print(f"tháng {thang} có 31 ngày")
    case 4|6|9|11:
        print(f"tháng {thang} có 30 ngày")
    case 2:
        if (nam % 4 == 0 and nam % 100 != 0) or (nam % 400 == 0):
            print(f"tháng {thang} có 29 ngày")
        else:
            print(f"tháng {thang} có 28 ngày")
    case _:
        print("tháng không hợp lệ")
```

## dùng match case kiểm tra chẵn lẻ

```python
n = int(input())
match n:
    case n if n % 2 == 0:
        print(f"{n} là số chẵn")
    case _:
        print(f"{n} là số lẻ")
```
