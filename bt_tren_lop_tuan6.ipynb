## kiểm tra mật khẩu có mạnh không với mật khẩu là 1 chuỗi có độ dài không nhỏ hơn 8, đủ các kí tự hoa, thường, chữ số, 1 trong các kí tự đặc biệt sau: @, #, !, $

```python
n = input()
while len(n) < 8:
    n = input("nhập mk gồm chữ hoa, chữ thường, số và kí tự đặc biệt: ")
chuthg = False
chuhoa = False
so = False
kitu = "@#!$"
ktdb = False
for i in n:
    if i.isupper() == True:
        chuhoa = True
    elif i.islower() == True:
        chuthg = True
    elif i.isdigit() == True:
        so = True
    elif i in kitu:
        ktdb = True
if chuhoa == chuthg == so == ktdb == True:
    print("mk mạnh")
else:
    print("mk yếu")
```
