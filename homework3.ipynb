## Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 

```python
h = int(input("Nhập độ cao h: "))
for i in range(1, h+1):
    for j in range(1, 2*h):
        if (j == h-i+1) or (j == h+i-1) or (i == h):
            print("*", end="")
        else:
            print(" ", end="")
    print() 
```

## Bài 6: Nhập vào số nguyên dương a, in ra tất cả các ước dương của a

```python
n = int(input("nhập 1 số nguyên dương:"))
ds = []
if n>0:
    for i in range(1,n+1):
        if n%i==0:
            ds.append(i)
    print("Các ước số của", n, "là:", " ".join(map(str, ds)))
else:
    print("không phải số nguyên dương")
```

## Bài 7: Nhập vào số nguyên dương a, đếm số ước dương của a 


```python
n = int(input("nhập 1 số nguyên dương:"))
ds = []
if n>0:
    for i in range(1,n+1):
        if n%i==0:
            ds.append(i)
    print(len(ds))
else:
    print("không phải số nguyên dương")
```

## Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b 

```python
a,b = map(int, input("nhập 2 số nguyên dương a và b: ").split())
ds = []
for i in range(1, min(a,b)+1):
    if a%i==0 and b%i==0:
        ds.append(i)
print(f"Ước chung của {a} và {b} là: {" ".join(map(str, ds))}")
```

## Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 


```python
n = int(input("Nhập một số nguyên dương: "))
if n>0:
    for i in range(1,n+1):
        if n/i==i:
            print(f"{n} là số chính phương")
            break
        else:
            print(f"{n} không phải là số chính phương")
            break
else:
    print("không phải số nguyên dương")
```

## Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 

```python
n = int(input("Nhập một số nguyên dương: "))
ds = []
for i in range(1,n):
    if n%i == 0:
        ds.append(i)
if sum(ds) == n:
    print(f"{n} là số hoàn hảo")
else:
    print(f"{n} không phải là số hoàn hảo")

```

## Bài 11: Nhập vào số nguyên a, nếu nhập số âm thì yêu cầu nhập lại cho đến khi người dùng nhập đúng số dương. Nếu người dùng nhập đúng số dương thì in ra “Bạn nhập đúng quy tắc” và dừng chương trình. 

```python
n = int(input("Nhập một số nguyên: "))
while True:
    if n>0:
        print("Bạn nhập đúng quy tắc")
        break
    else:
        print("Bạn nhập sai quy tắc, vui lòng nhập lại")
        n = int(input("Nhập một số nguyên: "))
```

## Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 


```python
n = int(input("Nhập một số nguyên dương: "))
s = 0
for i in range(1,n+1):
    s += i
    if s > n:
        print(s-i-(i-2))
        break
```
