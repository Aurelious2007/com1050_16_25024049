##  Nhập vào số nguyên dương a, in ra bảng cửu chương của a 

```python
n = int(input("nhập một số nguyên: "))
for i in range(1,11):
    print(f"{n} x {i} = {n*i}")
```

##  Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a,b = map(int, input("nhập hai số nguyên: ").split())
if a>b:
    for i in range(1,b+1):
        if a%i==0 and b%i==0:
            print(i)
elif a<b:
    for i in range(1,a+1):
        if a%i==0 and b%i==0:
            print(i)
else:
    print(a)
```

##  Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để in ra các số dương chẵn nhỏ hơn hoặc bằng n (n <= 1000) 

```python
n = int(input("nhập n: "))
while n<0 or n>1000:
    n = int(input("nhập n: "))
    for i in range(2,n+1,2):
        if n%2 == 0:
            print(i)
```

## Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input("nhập một số nguyên: "))
a = n
dem = 0
while n%2 == 0:
    n = n/2
    dem += 1
    if n%2 != 0:
        print(f"{a} = 2^{dem}")
        break
else:
    print("số bạn nhập không có dạng 2^k")
```

## Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("nhập một số nguyên dương: "))
tong = 0
while n>0:
    tong += n%10
    n = n//10
print(f"tổng các chữ số của số bạn nhập là: {tong}")
```

## Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input("nhập một số nguyên dương: "))
so_chan = 0
so_le = 0
while n>0:
    a = n%10
    n = n//10
    if a%2 == 0:
        so_chan += 1
    else:
        so_le += 1
print(f"số lượng chữ số chẵn: {so_chan}")
print(f"số lượng chữ số lẻ: {so_le}")
```

## Nhập n, Cho S(k) = 1 + 2 + 3 + … + k Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n  -->

```python
n = int(input("nhập một số nguyên: "))
if n%2 == 0:
    print(f"số k lớn nhất để S(k) < {n} là: {n//2 - 2}")
else:
    print(f"số k lớn nhất để S(k) < {n} là: {(n-1)//2}")
```
