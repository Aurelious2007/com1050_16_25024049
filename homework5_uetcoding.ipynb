## Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n.

```python
n = int(input())
if n < 2:
  print("no")
elif n == 2:
  print("2")
else:
  dem = 0
  ds = []
  for i in range(2,n+1):
    if n%i == 0:
      for k in range(2,int(i**0.5)+1):
        if i%k == 0:
          dem += 1
          break 
      if dem == 0:
         ds.append(i)
        
print(max(ds))
```

## Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.



```python
n = input()
dem = 0
while True:
    if n[::-1] == n:
        print(dem, n)
        break
    else:
        n = int(n) + int(n[::-1])
        n = str(n)
        dem += 1
    if dem > 100:
        print("stop")
        break
```

## Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number).



```python
n = int(input("nhập số nguyên dương:"))
ds = []
if n > 1:
    ds.append(1)
for i in range(2,n+1):
    for j in range(2,i+1):
        if (j**0.5)%1 == 0:
            if j not in ds:
                ds.append(j)
            else:
                continue
if ds == []:
    print("(no number)")
else:
    print(*ds)
        
```

- Nếu n là số lẻ: Số tiếp theo sẽ là 3n + 1.

## Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. Nếu có, in ra màn hình "true", ngược lại, in ra "false".

```python
a,b = map(int,input("nhập 2 số:").split())
uoc_a = []
uoc_b = []
for i in range(1,a):
    if a%i == 0:
        uoc_a.append(i)
for i in range(1,b):
    if b%i == 0:
        uoc_b.append(i)
if (sum(uoc_a) == b) and (sum(uoc_b) == a):
    print("true")
else:
    print("false")
```
